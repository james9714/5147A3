# Find top 15 amout postion and get the average salary

```python
data = pd.read_csv('salary.csv')
```

```python
data
```

```python
pos_sa = data.groupby('Position').mean()
```

```python
pos = data.groupby("Position").size()
```

```python
pos = pos.to_frame()
```

```python
pos = pos.rename(columns = {0:'Count'})
```

```python
pos.sort_values(by = ["Count"],ascending=False, inplace = True)
```

```python
pos_cnt_sa = pos.join(pos_sa)
```

```python
Top15 = pos_cnt_sa.round(0)[:15]
```

```python
in_list= list(range(1,16))
```

```python
Top15 = Top15.reset_index()
```

```python
top15pos = Top15['Position']
```

```python
Top15.to_csv('Top15position.csv',index=False)
```

# Work experience

```python
data1 = pd.read_csv('./IT Salary Survey EU  2020.csv')
```

```python
data2 = data1[['Total years of experience','Yearly brutto salary (without bonus and stocks) in EUR']]
```

```python
data2 = data2.rename(columns = {'Total years of experience':'Experience','Yearly brutto salary (without bonus and stocks) in EUR':'Salary'})
```

```python
data2 = data2.dropna(axis=0, how='any')
```

```python
data3=data2[data2['Experience'].apply(lambda x: str(x).isnumeric())]
```

```python
bins=[0,5,10,15,20,25,50]
labels = ["0-5","6-10","11-15","16-20","21-25","26+"]
```

```python
data3 = data3.astype(int)
```

```python
data3.dtypes
```

```python
data3["Seniority"] = pd.cut(x=data3["Experience"],bins=bins,labels=labels)
```

```python
data_exp = data3.groupby('Seniority')['Salary'].mean().round(0).to_frame().reset_index()
```

```python
data_exp
```

```python
pos_sal = data1[['Position ','Yearly brutto salary (without bonus and stocks) in EUR']]
```

```python
pos_sal  = pos_sal .rename(columns = {'Position ':'Position','Yearly brutto salary (without bonus and stocks) in EUR':'Salary'}).dropna(axis=0, how='any')
```

```python
pos_sal
```

```python
top15pos =top15pos.tolist()
```

```python
top15pos
```

```python
pos_sal_15 = pos_sal.loc[pos_sal['Position'].isin(top15pos)]
```

```python
pos_sal_15.describe()
```

```python
bins_sal=[10000,50000,60000,70000,80000,100000,400000]
labels_sal = ["10k-50k","50k-60k","60k-70k","70k-80k","80k-100k","100k+"]
```

```python
pos_sal_15["Salary_period"] = pd.cut(x=pos_sal_15["Salary"],bins=bins_sal,labels=labels_sal)
```

```python
pos_sal_15.sort_values(by = ['Salary'],inplace=True)
```

```python
df_sp = pos_sal_15.reset_index(drop=True)
```

```python
df_sp.groupby('Salary_period').size()
```

```python
dz = df_sp.groupby(['Salary_period','Position'], as_index=False).count()
```

```python
dz 
```

```python
da = pd.pivot(dz, index='Salary_period', columns='Position')
```

```python
da = da['Salary']
```

```python
da = da.reset_index()
```

```python
da.to_csv('Salary_line.csv',index=False)
```
